```python
import pandas as pd
```

```python
df_trips = pd.read_csv('DATARAW.csv', parse_dates=[0])
```

```python
df_trips.head()
```

```python
df_w = pd.read_csv('weathercleaned.csv', parse_dates=[0],index_col=0)
```

```python
df_w.head()
```

```python
# date columns in different format so going to split time into different column in weather df
df_w['datetime'] = pd.to_datetime(df_w['datetime']).dt.date
```

```python
df_w['HOUR'] = pd.to_datetime(df_w['datetime'].dt.hour)
```

```python
df_w.info()
```

```python
# change DAYOFSERVICE and TIME to datetime64
df_w['DAYOFSERVICE'] = pd.to_datetime(df_w['DAYOFSERVICE'])
df_w['HOUR'] = pd.to_datetime(df_w['HOUR']).dt.hour
```

```python
df_w.info()
```

```python
# create new column with ACTUALTIME_DEP in datetime format
df_trips['ACTUALTIME_DEP_HMS'] = pd.to_datetime(df_trips['ACTUALTIME_DEP'], unit='s')
```

```python
df_trips['HOUR'] = df_trips['ACTUALTIME_DEP_HMS'].dt.floor('h')
```

```python
df_trips['HOUR'] = pd.to_datetime(df_trips['HOUR']).dt.hour
```

```python
df_trips.info()
```

```python
df_trips = df_trips.merge(df_w, on=['DAYOFSERVICE','HOUR'])
```

```python
df_trips
```

drop rows which appended multiple weather values to each trip/day of service/hour match as per Concept Design Extract which states these variables act as the primary key.

```python
df_trips.drop_duplicates(subset=['TRIPID','HOUR','DAYOFSERVICE'], inplace=True)
```

```python
# drop unnecessary columns
df_trips = df_trips.drop(columns=['ACTUALTIME_DEP_HMS', 'dt_iso'])
```

```python
df_trips.reset_index(drop=True)
```
