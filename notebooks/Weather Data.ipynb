
import urllib.request
import sys

import csv
import codecs
        
try: 
  ResultBytes = urllib.request.urlopen("https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/setif/2023-03-01/2023-03-30?unitGroup=metric&include=hours&key=SUVGUFDZAVN55RSYQ7CHMDFFA&contentType=csv")
  
  # Parse the results as CSV
  CSVText = csv.reader(codecs.iterdecode(ResultBytes, 'utf-8'))
        
except urllib.error.HTTPError  as e:
  ErrorInfo= e.read().decode() 
  print('Error code: ', e.code, ErrorInfo)
  sys.exit()
except  urllib.error.URLError as e:
  ErrorInfo= e.read().decode() 
  print('Error code: ', e.code,ErrorInfo)
  sys.exit()

# Name of the CSV file to be saved #visualcrossing
filename = 'weathe2r.csv'

# Open the file in write mode and write the data
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    for row in CSVText:
        csvwriter.writerow(row)
